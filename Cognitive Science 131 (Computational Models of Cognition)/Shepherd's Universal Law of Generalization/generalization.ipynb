In this problem set, we are going to implement an analysis along the lines of Shepardâ€™s Universal Law
of Generalization. Imagine that you have one-dimensional stimuli. You should first construct a set of
10,000 consequential regions by randomly selecting a left edge in [-10,0] and a right edge in [0,10] so
that all of your regions contain x=0. It will be easiest if you store these consequential regions in a list

1.What is the probability that x=1 is in the concept (e.g. in any consequential region) given that
x=0 is?

```python
left_edge = np.random.uniform(low=-10, high=0, size=10000)
right_edge = np.random.uniform(low=0, high=10, size=10000)

edges = []
for i in range(len(left_edge)):
    edges.append([left_edge[i], right_edge[i]])
    
count = 0
x = 1
height_containsX = 0
total_height = 0

for i in range(len(edges)):
    total_height += 1/(edges[i][1]-edges[i][0])
    if x > edges[i][0] and x < edges[i][1]:
        count +=1
        height_containsX += 1/(edges[i][1]-edges[i][0])
    
probability = height_containsX/total_height
print(probability)
```

The probability is ~74%.

2.Plot the probability that x is in a consequential region as a function x. What does this
function look like? Write a sentence explaining why intuitively. 

This function looks like this because x = 0 will always be in the consequential region, and as x gets closer to each edge (-10 or 10) the probability decreases.

```python
x = []
y = []

for n in range(-100,101):
    x.append(n/10)
    heights_containsX = 0
    total_height = 0
    for j in range(10000):
        total_height += (1/(abs(edges[j][0]-edges[j][1])))
        if n/10 > edges[j][0] and n/10 < edges[j][1]:
            heights_containsX += (1/(abs(edges[j][0]-edges[j][1])))
    thingy = heights_containsX/total_height
    y.append(thingy)
```

```python
plt.plot(x,y)
plt.ylabel('Probability')
plt.xlabel('x')
plt.title('Probability that x is in a consequential region')
```

look for a straight line. Why does this check if the curve is exponential?

Plotting a logarithmic y axis and looking for a straight line checks if the curve is exponential because if you graph ln(e^x), the functions cancel each other out to make a straight line. So plotting a log y axis will produce a straight line if it cancels.
